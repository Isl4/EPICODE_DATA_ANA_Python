```python
##We have a list with the earnings of the last 12 months: 
# earnings = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]
#  using a for construct, calculate the average of the earnings 
# and print it on the screen

earnings= [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]
sum=0
for earning in earnings:
    sum += earning
    avg=sum/len(earnings)
    print(avg)
```
