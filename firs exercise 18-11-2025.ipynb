```python
fatturati_dict = {1997: 12_000, 1998: 15_000, 1999: 20_000, 2000: 23_000,
                 2001: 25_000, 2002: 17_000, 2003: 14_000, 2004: 21_000}
fatturati_series = pd.Series([12_000, 15_000, 20_000, 23_000, 25_000, 
                            17_000, 14_000, 21_000], 
                            index=range(1997, 2005)) 
fatturati_dict[1997] # this is a dict built-in function , it stocks items with easy access through key items
fatturati_series[1997]# this is a type of numpy/pandas.series, which like array. access is through tickets/labels 

#things we can do with series not possibile with dict --> statistic operations
fatturati_series.mean()
fatturati_series.max()
fatturati_series.min()
fatturati_series.std()# devision 
```
